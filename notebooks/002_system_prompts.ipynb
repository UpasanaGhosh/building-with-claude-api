```python
# Load env variables
from dotenv import load_dotenv
import os

print("API key loaded:", load_dotenv())
```

```python
# Create API client
from anthropic import Anthropic

client = Anthropic(api_key=os.getenv("ANTHROPIC_API_KEY"))
model = "claude-sonnet-4-0"
```

```python
# Helper Functions
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages, system=None):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
    }
    if system:
        params["system"] = system

    message = client.messages.create(**params)
    return message.content[0].text
```

## Friendly Maths Tutor

```python
# Without System Prompt
messages = []
add_user_message(messages, "How do I solve the equation 2x + 3 = 7?")
response = chat(messages)

response
```

```python
# With System Prompt
system_prompt = """
You are a patient math tutor.
Do not directly answer a student's questions.
Guide them to a solution step by step.
"""
response = chat(messages, system=system_prompt)

response
```

## Python Engineer

```python
messages = []
system_prompt = "You are a python engineer who writes very concise code"

add_user_message(messages, "Write a python function that checks a string for duplicate characters")
response = chat(messages=messages, system=system_prompt)

response
```
